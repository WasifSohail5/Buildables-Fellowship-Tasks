FIRST NOTEBOOK

Prompt Engineering Quiz

Question 1

What is prompting?
Answer: Prompting means providing instructions or guidance to an AI model so it knows how to respond to your request.

Question 2

In the following prompt, which part describes the primary task?
Prompt: “Explain how car batteries work using non-technical language that someone without any mechanical knowledge could understand and interpret.”
Answer: The main task is: “Explain how car batteries work in simple, non-technical terms.”

Question 3

When you are using a gen AI tool to create a healthy three-course menu that requires a certain number of calories, which prompting element will ensure the menu is more practical?
Answer: By setting clear constraints and giving sample inputs—like through one-shot prompting—the menu will be more realistic and usable.

Question 4

Prompting with audio to receive a visual representation is called what?
Answer: This is known as multimodal prompting.

Question 5

What does a human-in-the-loop approach to prompting emphasize?
Answer: It highlights the role of a human reviewer who verifies, adjusts, and ensures that the AI’s responses are accurate, safe, and ethically appropriate.

Question 6

When the initial prompt "Create a dataset" is too broad, what should you do to refine it?
Answer: You should make it more specific by including details such as the dataset’s subject, required size, structure, and preferred output format.


SECOND NOTEBOOK

Exercises: Try It Yourself

Exercise 1: Icebreakers

Write a prompt that asks AI to create 3 fun icebreakers for a new teammate who enjoys basketball and photography.

Prompt:
"You are a clever conversation starter with a knack for humor. Suggest 3 quirky icebreakers for my new teammate that can spark instant laughter and break the tension. (They’re into basketball and photography)."

Exercise 2: Product Launch

Imagine you’re introducing a new mobile fitness app (or any product). Write a prompt to:

Outline a 6-month launch timeline.

Suggest 3 pre-launch buzz strategies.

Prompt:
"You are a creative marketing strategist. Task: Build a 6-month launch plan for a brand-new mobile fitness app. Context: The app turns fitness goals into gamified quests, similar to video games. Format: Provide (1) a fun month-by-month roadmap, and (2) 3 wild pre-launch buzz stunts that would get people hyped on social media."

Example buzz ideas:

“TikTok challenge: 30-second dance-off with a custom soundtrack.”

“Mystery fitness loot boxes shipped to influencers.”

“An AR teaser game where users uncover hidden workouts in their city.”

Exercise 3: Global Audience Twist

Take your product launch prompt and ask the AI to modify it for a worldwide audience.

Prompt:
"You are a cultural trend hacker. Task: Adapt the 6-month fitness app launch campaign for global reach. Context: The ideas should feel inclusive and culturally playful. Format: Present the ideas in a table that shows differences across North America, Europe, and Asia. Keep the tone fun and make each region’s buzz ideas unique."

Exercise 1: Project Tracker

Prompt:
"You are a smart project assistant. Create a project tracker in table form with the columns: Task, Owner, Deadline, and Status. Constraint: No owner should manage more than 2 active tasks at once. Provide at least 6 sample tasks with realistic deadlines and statuses (e.g., Not Started, In Progress, Completed)."

Exercise 2: Event Schedule

Prompt:
"You are an event planner. Build a 3-day conference schedule in a table with the columns: Day, Session Title, Speaker, and Time. Constraint: Each speaker can only present once per day. Titles should sound engaging, and times should not overlap."

Exercise 3: Team Availability (Vacation Tracker)

Prompt:
"You are an HR assistant. Create a vacation tracker for 8 employees with columns: Employee, Vacation Start, Vacation End. Constraint: At least 5 employees must remain available at all times (so no more than 3 on vacation simultaneously). Distribute the vacations realistically across one month."

Reflection Question
How could you use AI to organize recurring tasks (like meetings, deadlines, or resources) at your workplace?

Answer: AI could help by:

Auto-creating schedules for meetings and deadlines.

Balancing workloads so no one gets overloaded.

Tracking resources like rooms, equipment, and task owners.

Sending smart reminders for repeating tasks.

Exercise 1: Speaker-based Minutes

Prompt:
"Using a sample transcript (or your last meeting notes), summarize the meeting into minutes organized by speaker. Add 2–3 bullet points per speaker and clearly highlight action items."

Exercise 2: Action-Focused Notes

Prompt:
"Write meeting minutes that capture only the action items. Format each entry as: Action Item, Responsible Person, Deadline."

Exercise 3: Thematic Notes

Prompt:
"Organize meeting minutes by themes or topics. For each section, summarize the key decisions and list any assigned action items."

Reflection Question
How could AI-generated meeting notes help improve follow-up and accountability at work?

Answer: AI can:

Capture all action items so nothing is forgotten.

Assign clear ownership for each task.

Speed up follow-ups by turning discussions into to-do lists.

Boost accountability since responsibilities and deadlines are documented instantly.

Advanced Prompting (CoT, ToT, Meta-Prompting)

Reflection Question
How could CoT and ToT prompting help you solve complex problems or spark creativity in your workplace?

Answer:

Chain of Thought (CoT) encourages step-by-step reasoning.

Tree of Thoughts (ToT) explores multiple solution paths before choosing one.

Together, they help break down complex issues and encourage creative thinking rather than rushing into a single answer.

Reflection Question
How could meta-prompting help you craft better prompts or improve AI outputs for your specific work tasks?

Answer: Meta-prompting acts like a coach for writing prompts. It helps me ask clearer questions, reuse effective prompts, and improve outputs consistently without extra effort.
